```python
# Grouping data together

import pandas as pd
```

```python
data_import = pd.read_csv("data_cleaning_challenge.csv")
data_import
```

```python
drop_extra = data_import.drop(columns=["Unnamed: 9", "Unnamed: 10"])
drop_extra
```

```python
drop_nans = drop_extra[drop_extra["Row Type"].notna()]
drop_nans
```

```python
# Iterate thru to make new columns representing group number. 
column_values = []
counter = 0

for i in drop_nans["Row Type"]:
    if "first name" in i:
        counter += 1
    column_values.append(counter)
```

```python
column_values
```

```python
#Include the values into a new column
iter_cols = drop_nans
iter_cols["Iteration"] = column_values
iter_cols
```

```python
# Getting rid of extra header name
drop_extra_column_name = iter_cols[iter_cols["Row Type"] != "Row Type"]
drop_extra_column_name
```

```python
name_dataframe = drop_extra_column_name[drop_extra_column_name["Row Type"].str.contains("first name")]
name_dataframe.drop(columns=["Speed1", "Speed2", "Electricity", "Weight", "Torque"], inplace=True)
name_dataframe.rename(columns = {"Row Type": "First Name", "Iter Number": "Last Name", "Power1": "Date"}, inplace = True)
name_dataframe["First Name"] = name_dataframe["First Name"].str[12:]
name_dataframe["Last Name"] = name_dataframe["Last Name"].str[11:]
name_dataframe["Date"] = name_dataframe["Date"].str[6:]
name_dataframe
```

```python
no_name_dataframe = drop_extra_column_name[~drop_extra_column_name["Row Type"].str.contains("first name")]
no_name_dataframe
```

```python
pd.merge(left=name_dataframe, right=no_name_dataframe, how="inner", on="Iteration")
```
